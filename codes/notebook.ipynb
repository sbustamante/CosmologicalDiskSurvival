```python
#==================================================================================================
#Importing all necessary libraries
#==================================================================================================
%pylab inline
from IPython.display import HTML
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# JSAnimation import available at https://github.com/jakevdp/JSAnimation
from JSAnimation import IPython_display
from matplotlib import animation
import matplotlib.image as mpimg
import numpy as np
import os
```

```python
#==================================================================================================
#Parameters
#==================================================================================================
#Web scheme to use
web = "Tweb"
#Resolution
res = 256
#Threshold value
l_thr = 0.265
#Box Lenght
Lbox = 250.

#Local data
datafold = "../data/"
#Big data of Bolshoi
bolshoidata = "../../CosmicData/BOLSHOI/%s/%d/"%(web, res)
```

```python
#==================================================================================================
#General functions
#==================================================================================================
#Classification Scheme
def Scheme( eig1, eig2, eig3, Lamb ):
    N1 = len( eig1 )
    try:
        N2 = len( eig1[0] )
        sch = 0*eig1
        for i in xrange(N1):
            for j in xrange(N2):
                if eig1[i,j] > Lamb: sch[i,j] += 1
                if eig2[i,j] > Lamb: sch[i,j] += 1
                if eig3[i,j] > Lamb: sch[i,j] += 1
        return sch
    except:
        sch = 0*eig1
        for i in xrange(N1):
            if eig1[i] > Lamb: sch[i] += 1
            if eig2[i] > Lamb: sch[i] += 1
            if eig3[i] > Lamb: sch[i] += 1
        return sch
        

#Cut Field
def CutFieldZ( filename, X, res=32, Coor = 3, N=256 ):
    '''
    Coor  1 -- X      2 -- Y      3 -- Z
    '''
    if res == 'plain':
	os.system( "./Field_Cut_Plain.out %s %d temp.tmp %d %d"%( filename, X, Coor, N ) )
	datos = np.loadtxt( 'temp.tmp' )
	N = int(np.sqrt(len( datos )))
	datos = datos.reshape( (N, N) )
	os.system( "rm temp.tmp" )
      
    else:	
	os.system( "./Field_Cut%d.out %s %d temp.tmp %d"%( res, filename, X, Coor ) )
	datos = np.loadtxt( 'temp.tmp' )
	N = int(np.sqrt(len( datos )))
	datos = datos.reshape( (N, N) )
	os.system( "rm temp.tmp" )
    return datos

#Load data for a given cell
def Loader( filename, X, Y, Z, res=32 ):
    try:
        N = len(X)
        field = np.zeros(N)
        for ic in xrange(N):
            i = int(np.floor(X[ic]/Lbox*res))
            j = int(np.floor(Y[ic]/Lbox*res))
            k = int(np.floor(Z[ic]/Lbox*res))
            datos = np.transpose(CutFieldZ( filename, i, res=32, Coor = 1 ))
            field[ic] = datos[ j, k ]
        return field
    except:
        i = int(np.floor(X/Lbox*res))
        j = int(np.floor(Y/Lbox*res))
        k = int(np.floor(Z/Lbox*res))
        datos = np.transpose(CutFieldZ( filename, i, res=32, Coor = 1 ))
        return datos[ j, k ]
```

## Isolated Pairs

Local Group candidates in Bolshoi are going to be selected from the general fof catalogue of dark matter halos following the next criteria:

- Each halo has a mass in the range $7\times 10^{11} M_{\odot}<M_{h}<7\times 10^{12} M_{\odot}$, consistent with the estimated values for MW and M31.
- Regarding each halo, there cannot be any other halo within the mass range $7\times 10^{11} M_{\odot}<M_{h}<7\times 10^{12} M_{\odot}$ closer than its companion.
- The relative radial velocity between both halos must be negative.
- The distance between the centre of mass of the halos must be less than $1.0\ Mpc$.
- There cannot be halos more massive than $7\times 10^{12} M_{\odot}$ within a radius of $3\ Mpc$ to every object centre.
- There cannot be halos more massive than $7\times 10^{13} M_{\odot}$ within a radius of $4\ Mpc$ to every object centre (Virgo-like systems).

For this selection we are going to use the code **[Pair-Finder](https://github.com/sbustamante/Pair-Finder)**. This code produces a catalogue of general pairs (i.e. pairs of halos that are the closest one to each other), and an isolated sample with all the previous criteria. Next is shown the used parameters file:

```python
#Loading produced Isolated-Pair catalogue (id1,M1,id2,M2,rrel,vrad,vtan)
isopairs = np.loadtxt( "%s/isolated_pairs_fof.dat"%(datafold), usecols=(1,2,4,5,7,8,9) )
#Number of halos
Nisop = len(isopairs)
#Properties of halo1 of each isolated pair
halos1 = halos[ (isopairs[:,0]-1).astype(int) ]
#Properties of halo2 of each isolated pair
halos2 = halos[ (isopairs[:,2]-1).astype(int) ]
```

```python
#Plotting a slide
Zmin = 0
Zmax = 10
X = halos[ (halos[:,3]>=Zmin)&(halos[:,3]<=Zmax), 1 ]
Y = halos[ (halos[:,3]>=Zmin)&(halos[:,3]<=Zmax), 2 ]
X1 = halos1[ (halos1[:,3]>=Zmin)&(halos1[:,3]<=Zmax), 1 ]
Y1 = halos1[ (halos1[:,3]>=Zmin)&(halos1[:,3]<=Zmax), 2 ]
X2 = halos2[ (halos2[:,3]>=Zmin)&(halos2[:,3]<=Zmax), 1 ]
Y2 = halos2[ (halos2[:,3]>=Zmin)&(halos2[:,3]<=Zmax), 2 ]


plt.figure( figsize=(8,8) )
plt.plot( X, Y, ",", color="black" )
plt.plot( X2, Y2, "*", color="blue", alpha = 1.0, ms = 10 )
plt.plot( X1, Y1, "o", color="red", alpha = 0.5, ms = 7 )
plt.xlabel( "X [$h^{-1}Mpc$]", fontsize=14 )
plt.ylabel( "Y [$h^{-1}Mpc$]", fontsize=14 )
plt.title( "Isolated systems in Bolshoi", fontsize=14 )
```

## LG criteria

Besides the previous criteria, we also propose two extra constraints based on the observational reported values of the tangential and radial velocities:

```python
#Radial velocity km/s
vrad = -109.0
dvrad = 4.4
#Tangential velocity km/s
vtan = 34.4
```

```python
#Sample of LG like systems
LGpairs = isopairs[ (isopairs[:,5]<=vrad+dvrad)&(isopairs[:,5]>=vrad-dvrad)&(isopairs[:,6]<=vtan) ]
#Number of LG systems
Nlg = len(LGpairs)
#Properties of halo1 of each isolated pair
lghalos1 = halos[ (LGpairs[:,0]-1).astype(int) ]
#Properties of halo2 of each isolated pair
lghalos2 = halos[ (LGpairs[:,2]-1).astype(int) ]

print Nlg, "LG systems"
```

```python
#Plotting environment of Bolshoi
my_cmap4 = plt.cm.get_cmap('gray', 4)
#Loading density and eigenvalues
delta_filename = '%s/Delta_s1'%(bolshoidata)
#Loading Vweb filename
eig_filename = '%s/Eigen_s1'%(bolshoidata)

#Coordinate to cut (1 -- X, 2 -- Y, 3 -- Z)
axe = 3
#Cut
Cut = 5

#Loading Fields
delta = CutFieldZ( delta_filename, Cut, 32, Coor = axe )
eig1 = CutFieldZ( eig_filename+"_1", Cut, 16, Coor = axe )
eig2 = CutFieldZ( eig_filename+"_2", Cut, 16, Coor = axe )
eig3 = CutFieldZ( eig_filename+"_3", Cut, 16, Coor = axe )
```

```python
plt.figure( figsize=(8,8) )
plt.imshow( -Scheme( eig1, eig2, eig3, l_thr )[:,::-1].T, extent = [0,Lbox,0,Lbox], vmin=-3, vmax=0, cmap = my_cmap4 )
plt.plot( X, Y, ",", color="blue" )
plt.plot( X2, Y2, "*", color="red", alpha = 1.0, ms = 10, label="Isolated pairs" )
plt.xlabel( "X [$h^{-1}Mpc$]", fontsize=14 )
plt.ylabel( "Y [$h^{-1}Mpc$]", fontsize=14 )
plt.title("Environment of isolated pairs", fontsize=14 )
plt.legend( )
```
